CREATE DATA FRAME FROM DICTIONARY

```python
weather_data = {
    'day': ['1/1/2026','1/2/2026','1/3/2026'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain','Sunny','Snow']
}
```

```python
df = pd.DataFrame(weather_data)
df
```

DF FROM TUPLE

```python
#From Tuple
weather_data = [
    ('1/1/2026',32,6,'Rain'),
    ('1/2/2026',35,7,'Sunny'),
    ('1/3/2026',28,2, 'Snow')
]

df = pd.DataFrame(weather_data, columns=['day','temperature','windspeed','event'])
df
```

DF FROM LIST OF DICTIONARIES

```python
#List of Dictionaries+

weather_data = [
    {'day': '1/1/2026', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2026', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2026', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
]
df = pd.DataFrame(weather_data)
df
```
