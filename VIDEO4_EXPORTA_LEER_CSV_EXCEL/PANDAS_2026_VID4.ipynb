```python
df = pd.read_csv('stock_data.csv', na_values={
    'eps':['not available','n.a.'],
    'revenue':['not available','n.a.',-1],
    'people':['not available','n.a.'],
    'price':['not available','n.a.']
})
df
```

CREATING TWO DATAFRAMES FOR AN EXAMPLE

```python
import pandas as pd
import numpy as np

# DataFrame 1
data1 = {
    'col_A': np.random.randint(1, 100, 5),
    'col_B': np.random.rand(5).round(2),
    'col_C': ['apple', 'banana', 'orange', 'grape', 'kiwi']
}
df_random1 = pd.DataFrame(data1)
display(df_random1)
```

```python
# DataFrame 2
data2 = {
    'item': ['shirt', 'pants', 'shoes', 'hat'],
    'price': [25.99, 45.50, 70.00, 15.25],
    'in_stock': [True, False, True, True]
}
df_random2 = pd.DataFrame(data2)
display(df_random2)
```

WRITING 2 DF IN ONE EXCEL

```python
with pd.ExcelWriter('output.xlsx') as writer:
  df_random1.to_excel(writer, sheet_name='Sheet1', index=False)
  df_random2.to_excel(writer, sheet_name='Sheet2', index=False)
```
