READ CSV AND FORMATTING THE DATES

```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df
```

CHEKING THE DATE FORMAT

```python
type(df.day[0])
```

SETTING THE DAY AS AN INDEX

```python
df.set_index('day', inplace=True)
df
```

FILLNA WITH ZERO

```python
new_df = df.fillna(0)
new_df
```

FILLNA WITH DIFFERENT VALUES

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})

new_df
```

FILLNA WITH FORWARD FILL

```python
new_df = df.fillna(method='ffill')
new_df
```

BACKWARD FILL

```python
new_df = df.fillna(method='bfill')
new_df
```

BACKWARD FILL WITH AXIS COLUMNS

```python
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

FORWARD FILL WITH JUST ONE FORWARD

```python
new_df = df.fillna(method='ffill', limit=1)
new_df
```

COMPLETE NA WITH INTERPOLATE

```python
new_df = df.interpolate()
new_df
```

INTERPOLATE BASED ON TIME

```python
new_df = df.interpolate(method='time')
new_df
```

DROPNA

```python
new_df = df.dropna()
new_df
```

DROPNA WITH ALL (JUST DROP THE COLUMNS WITH ALL NA VALUES)

```python
#Dropear las filas que tenga todos sus valores vacios
new_df = df.dropna(how='all')
new_df
```

KEEP THE ROWS WITH AT LEAST ONE NO N.A. VALUE

```python
#At least 1 no N.A. value keep it
new_df = df.dropna(thresh=1)
new_df
```

```python
#I need at least 2 no N.A. Values
new_df = df.dropna(thresh=2)
new_df
```

CREATING NEW INDEX TO FILL THE MISSING DATES

```python
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

Esta línea crea un rango de fechas. `pd.date_range()` es una función de pandas que genera una secuencia de objetos `datetime`. En este caso, crea un rango de fechas que comienza el '01-01-2017' y termina el '01-11-2017' (ambos incluidos), con una frecuencia diaria por defecto.

**`dt = pd.date_range('01-01-2017', '01-11-2017')`**

Aquí, la secuencia de fechas `dt` que creaste en el paso anterior se convierte explícitamente en un objeto `pd.DatetimeIndex`. Aunque `date_range` ya suele devolver un DatetimeIndex, esta línea asegura que el tipo de índice sea el correcto.

**`idx = pd.DatetimeIndex(dt)`**

Finalmente, esta es la parte crucial. El método `reindex()` se usa para conformar un DataFrame a un nuevo índice. Lo que hace es lo siguiente:

1.  **Añade filas:** Si el nuevo índice (`idx`) contiene fechas que no estaban en el índice original de `df`, `reindex()` añade esas filas al DataFrame. Para los valores de las columnas en estas nuevas filas, pandas inserta `NaN` (Not a Number) o `NaT` (Not a Time) para indicar que no hay datos disponibles para esas fechas.
2.  **Mantiene filas:** Si una fecha del nuevo índice ya existía en el DataFrame original, los datos correspondientes a esa fecha se mantienen.
3.  **Elimina filas:** Si el DataFrame original tenía fechas en su índice que no están presentes en el nuevo índice `idx`, esas filas se eliminan del DataFrame resultante.

En resumen, estás creando un rango de fechas completo y luego le estás diciendo a tu DataFrame `df` que se "ajuste" a ese nuevo rango de fechas, rellenando los días faltantes con `NaN` si no había datos para ellos en el `df` original.

**`df = df.reindex(idx)`**

El método `time` utiliza los valores del índice de tiempo para estimar los valores faltantes. Por ejemplo, si tienes valores en el día 1 y el día 3, y falta un valor en el día 2, `time` considerará la distancia temporal entre estos puntos para hacer la interpolación, en lugar de simplemente la posición lineal. Esto es útil cuando los datos no están espaciados uniformemente en el tiempo.

En tu caso, después de ejecutar `df.set_index('day', inplace=True)`, el índice de tu DataFrame `df` se convirtió en un `DatetimeIndex`. Por lo tanto, cuando llamaste a `df.interpolate(method='time')`, pandas utilizó automáticamente este `DatetimeIndex` para realizar la interpolación basada en el tiempo.

Si el índice no fuera de tipo `DatetimeIndex`, `method='time'` generaría un error, ya que no tendría información temporal para basar la interpolación.
